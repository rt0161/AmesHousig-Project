```python
# evaluate by 10 fold cv
from sklearn.model_selection import StratifiedKFold, cross_val_score
# define 10-fold cross validation test harness
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
cvscores = []
for train, test in kfold.split(train_data, pc_train):
  # create model
    model = Sequential()
    model.add(Dense(13,input_dim=13, kernel_initializer='normal', activation='relu'))
    model.add(Dense(12, kernel_initializer='normal',activation='relu'))
    model.add(Dense(11, kernel_initializer='normal',activation='relu'))
    model.add(Dense(10, kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error',optimizer = 'adam')
    # Fit the model
    model.fit(train_data.loc[train], pc_train[train], epochs=1800, verbose=0)
    # evaluate the model
    scores = model.evaluate(train_data.loc[test], pc_train[test], verbose=0)
    print("%s: %.2f" % ('MSE', np.sqrt(scores)))
    cvscores.append(np.sqrt(scores))
```

```python
## SAVE the cv scores , for time sake
with open('KNN_Keras_cv_scoresoutput.txt', 'w') as output_file:
    for value in cvscores:
        output_file.write(str(value))
```

```python
# combine all data
pric_all = pc_train.append(pc_test)
full = train_data.append(test_data)
full['SalePrice']=pric_all.values

# Boostrap estimate RF model confidence interval:
n_iterations = 1000
n_size = int(len(full) * 0.50)

values=full.values
```

```python
from sklearn.utils import resample
from sklearn.ensemble import RandomForestRegressor
stats = list()
for i in range(n_iterations):
    # prepare train and test sets
    train = resample(values, n_samples=n_size)
    test = np.array([x for x in values if x.tolist() not in train.tolist()])
    # fit model
    RF_regr = RandomForestRegressor(random_state=1)
    RF_regr.fit(train[:,:-1], train[:,-1])
    # evaluate model
    predictions = RF_regr.predict(test[:,:-1])
    score = np.sqrt(metrics.mean_squared_error(test[:,-1], predictions))
    print(i,"th iteration....score =",score)
    stats.append(score)
```

```python
## SAVE the cv scores , for time sake
with open('RF_Sklearn_bstp_mseoutput.txt', 'w') as output_file:
    for value in stats:
        output_file.write(str(value))
```

```python

# plot boostrap mse scores
fig, ax = plt.subplots(figsize=(10,8))
plt.hist(stats,bins=100)
# confidence intervals

alpha = 0.95
p = ((1.0-alpha)/2.0) * 100
lower = np.percentile(stats, p)
p = (alpha+((1.0-alpha)/2.0)) * 100
upper = np.percentile(stats, p)
plt.plot([lower, lower],[0,40],'r--')
plt.plot([upper, upper],[0,40],'r--')
plt.show()
print('%.1f confidence interval %.1f and %.1f' % (alpha*100, lower, upper))

```
