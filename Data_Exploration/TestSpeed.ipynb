```python
import pandas as pd
data_all = pd.read_csv('AmesHousing_condi.csv')
```

```python
data_all.pop('Unnamed: 0')
data_all.head()
```

```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
# create baseline model
model = Sequential()
model.add(Dense(13,input_dim=13, kernel_initializer='normal', activation='relu'))
model.add(Dense(12, kernel_initializer='normal',activation='relu'))
model.add(Dense(11, kernel_initializer='normal',activation='relu'))
model.add(Dense(10, kernel_initializer='normal',activation='relu'))
model.add(Dense(1,kernel_initializer='normal'))
```

```python
# compile model
model.compile(loss='mean_squared_error',optimizer = 'adam')
```

```python
train_data = data_all[0:1700]
pc_train= train_data['SalePrice']
train_data = train_data.drop('SalePrice', axis =1)
```

```python
import keras.callbacks
from time import time
class TimeHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.times = []

    def on_epoch_begin(self, batch, logs={}):
        self.epoch_time_start = time.time()

    def on_epoch_end(self, batch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)
```

```python
import time
start_time = time.time()
history=model.fit(train_data,pc_train,epochs=1800)
print(time.time() - start_time, "seconds")
```

```python
# Try out 10-fold cv on Keras and access timing
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
# define 10-fold cross validation test harness
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
cvscores = []
start_time = time.time()
for train, test in kfold.split(train_data, pc_train):
  # create model
    model = Sequential()
    model.add(Dense(13,input_dim=13, kernel_initializer='normal', activation='relu'))
    model.add(Dense(12, kernel_initializer='normal',activation='relu'))
    model.add(Dense(11, kernel_initializer='normal',activation='relu'))
    model.add(Dense(10, kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error',optimizer = 'adam')
    # Fit the model
    model.fit(train_data.loc[train], pc_train[train], epochs=1800, verbose=0)
    # evaluate the model
    scores = model.evaluate(train_data.loc[test], pc_train[test], verbose=0)
    print("%s: %.2f" % ('MSE', np.sqrt(scores)))
    cvscores.append(np.sqrt(scores))
print(time.time() - start_time, "seconds")
```
